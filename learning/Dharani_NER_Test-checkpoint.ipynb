# Named Entity Recognition
1. NLP task to Identify important named entities in the text
    Can be people, places,Dates etc..


```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
text = 'In New York, I like to ride Subway to visit Statue of Liberty and some restaurants rated well by John and Kate'
```

```python
print(text)
```

```python
words  = nltk.word_tokenize(text)
```

```python
words
```

```python
tagged_words = nltk.pos_tag(words)
```

```python
tagged_words
```

```python
print(nltk.ne_chunk(tagged_words))
```

## NER with NLTK using scraped news article

```python
article= open('C:\\Users\\Vandana\\Desktop\\Ndtv.txt','r').read()
```

```python
sentences = nltk.sent_tokenize(article)
```

```python
sentences
```

```python
len(sentences)
```

```python
word_sentence = [nltk.word_tokenize(s) for s in sentences ]
```

```python
word_sentence
```

```python
pos_sentence = [nltk.pos_tag(st) for st in word_sentence]
```

```python
pos_sentence
```

```python
chunk_sentence = nltk.ne_chunk_sents(pos_sentence,binary = True)
type(chunk_sentence)
```

```python
chunk_sentence
```

```python
for sent in chunk_sentence:
    for chunk in sent:
        if hasattr(chunk, "label") and chunk.label() == "NE":
            print(chunk)
```
